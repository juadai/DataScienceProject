```python
import pandas as pd
import numpy as np

data = pd.read_excel("CWD assessment data all years.xlsx")
data
```

# Take total CWD and the treatment applied as the only data 

```python
col_names = np.array(data.iloc[1])[1:]
treatment = ["Gap", "Control", "Radial", "Control", "Radial", "Control", "Radial", "Gap", "Radial", "Gap", "Control", "Gap"]
yr0 = np.array(data.iloc[5])[1:]
yr6 = np.array(data.iloc[14])[1:]

##add fenced or not & type



total = pd.DataFrame(data = [treatment, yr0, yr6], index = ["treatment", "year 0", "year 6"],columns = col_names)


total
```

```python
result = total.T
result["difference"] = result["year 6"] - result["year 0"]
result["stumps_diff"] = result["stumps6"] - result["stumps0"]
result
```

```python
result.to_csv("output.csv")
```
