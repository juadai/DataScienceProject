#### remove useless columns

```python
# remove useless columns
df_understorey = df_understorey.drop(columns = ['Lifeform', 'Percentage_Avg', 'Percentage_max', 'Percentage_min', 'Field_code'])
df_understorey.tail()
```

#### From the summary above, there are some small mistakes which need to be fixed
1. miss-spelling of column name: Quardat_fenced, should be: Quadrat_fenced
2. "corrrect" in Quadrat_Can_Cov should be "correct"
3. "Control" in Quadrat_gap_or_no should be "control"
4. duplicates in Record_ID
5. need further check of T002_Flora_Species_name attribute

#### fix 1.

```python
# previously: 'Quardat_fenced'
correct_column = list(df_understorey.columns)
correct_column[correct_column.index('Quardat_fenced')] = 'Quadrat_fenced'
df_understorey = df_understorey.rename(columns=dict(zip(list(df_understorey.columns), correct_column)))
df_understorey.columns
```

#### fix 2.

```python
# previously: correct=2499, corrrect=33
df_understorey.loc[df_understorey['Quadrat_Can_Cov'] == 'corrrect', 'Quadrat_Can_Cov'] = 'correct'
df_understorey['Quadrat_Can_Cov'].value_counts()
```

#### fix 3.

```python
# previously: control=448, Control=158
df_understorey.loc[df_understorey['Quadrat_gap_or_no'] == 'Control', 'Quadrat_gap_or_no'] = 'control'
df_understorey['Quadrat_gap_or_no'].value_counts()
```

#### fix 4.

```python
# counts of duplicate 'Record_ID', eg: 982 occurrence of 2 rows having same Record_ID
(df_understorey['Record_ID'].value_counts()).value_counts()
```

```python
# the duplicate Record_ID values
s_recordid = df_understorey['Record_ID'].value_counts() != 1
duplicate_id = s_recordid.loc[s_recordid.values == True].index
duplicate_id
```

```python
# remove duplicate rows (keep the first duplicate rows)
for i in duplicate_id:
    index = list(df_understorey.loc[df_understorey['Record_ID'] == i].index)
    df_understorey = df_understorey.drop(index = index[1:])
```

### The following code checks whether duplicate rows are really duplicate (i.e. whether rows with same 'Record_ID' have the same values for all other attributes)

Result from the checking: Except for the 'Lifeform' attribute, rows with same 'Record_ID' have the same values for all other attributes. As 'Lifeform' is an attribute to be removed anyway, then rows with same 'Record_ID' are really duplicate rows. So it's ok to remove duplicates as in the code above.

Not necessary to run the code, it will first load the data into another dataframe and process as before.

```python
df_check = pd.read_excel("../datasets/Ecothining_understoreydata2020.xlsx")
```

```python
correct_column = list(df_check.columns)
correct_column[correct_column.index('Quardat_fenced')] = 'Quadrat_fenced'
df_check = df_check.rename(columns=dict(zip(list(df_check.columns), correct_column)))
df_check.loc[df_check['Quadrat_Can_Cov'] == 'corrrect', 'Quadrat_Can_Cov'] = 'correct'
df_check.loc[df_check['Quadrat_gap_or_no'] == 'Control', 'Quadrat_gap_or_no'] = 'control'
```

```python
s_recordid_check = df_check['Record_ID'].value_counts() != 1
duplicate_id_check = s_recordid_check.loc[s_recordid_check.values == True].index
duplicate_id_check
```

```python
# check whether rows with same 'Record_ID' have the same values for all other attributes
same = 0
notsame = 0
for i in duplicate_id_check:
    index = list(df_check.loc[df_check['Record_ID'] == i].index)
    first = df_check.iloc[index[0]]
    same += 1
    for j in index[1:]:
        current = df_check.iloc[j]
        if (current.equals(first)):
            same += 1
        else:
            print(index[0], j, i)
            notsame += 1

print(same, notsame) # result: 13195 rows are duplicate rows, 166 "duplicate" rows have different other attribute values
```

```python
# notice: 166 is the number of 'Lifeform' being not NaN
df_check['Lifeform'].value_counts()
```

```python
# take one pair of notsame rows to check
df_check.iloc[[11860, 11861]]
```

```python
# check whether rows with same 'Record_ID' have the same values for all other attributes, eliminating impact of 'Lifeform'
same = 0
notsame = 0
for i in duplicate_id_check:
    index = list(df_check.loc[df_check['Record_ID'] == i].index)
    first = df_check.iloc[index[0], :-1]
    same += 1
    for j in index[1:]:
        current = df_check.iloc[j, :-1]
        if (current.equals(first)):
            same += 1
        else:
            print(index[0], j, i)
            notsame += 1

print(same, notsame) # result: when eliminating 'Lifeform', all duplicate rows are really duplicate
```

#### Result:
Except for the 'Lifeform' attribute, rows with same 'Record_ID' have the same values for all other attributes. As 'Lifeform' is an attribute to be removed anyway, then rows with same 'Record_ID' are duplicate rows. Now we can remove duplicates.
